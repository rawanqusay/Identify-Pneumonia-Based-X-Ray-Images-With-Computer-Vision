```python
# Import libraries
import pickle
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image
```

```python
# Connecting Google Colab with Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load Model
model = load_model('/content/drive/MyDrive/model_tl.h5')
```

```python
# Predict Inference Data
file_path = "/content/drive/MyDrive/NORMAL-831813-0001.jpeg"

if os.path.exists(file_path):
    img = tf.keras.utils.load_img(file_path, target_size=(224, 224))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    x = tf.keras.utils.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    # melakukan predict
    classes = model.predict(x)
    result_pred = np.where(classes < 0.7, 0, 1)
    if result_pred == 1:
        print('Pneumonia')
    else:
        print('Normal')
else:
    print("File not found:", file_path)
```
